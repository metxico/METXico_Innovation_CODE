```python
import pandas as pd
import plotly.express as px

# Cargar el archivo CSV
df = pd.read_csv('InputOFfilename.csv')

# Verificar las columnas
print(df.columns)

# Contar cuántas veces aparece cada categoría en 'ClassyFire_Superclass'
grouped_data = df['ColumnNameOFinterest'].value_counts()

# Crear un gráfico de pie interactivo
fig = px.pie(values=grouped_data, names=grouped_data.index, title='OutPutOFpiechart')

# Mostrar el gráfico interactivo
fig.show()
```
