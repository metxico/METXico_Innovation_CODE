```python
import umap
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv('Inputfilename.csv')
```

```python
!pip install rdkit-pypi
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
molecules = [Chem.MolFromSmiles(smiles) for smiles in df['Smiles']]
descriptors = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024) for mol in molecules]
descriptors_array = np.array([list(desc) for desc in descriptors])
```

```python
umap_3d = umap.UMAP(n_components=3)
reduced_descriptors_3d = umap_3d.fit_transform(descriptors_array)
```

```python
df_reduced = pd.DataFrame(reduced_descriptors_3d, columns=['UMAP1', 'UMAP2', 'UMAP3'])
df_reduced['Library'] = df['Library']
```

```python
fig = px.scatter_3d(df_reduced, x='UMAP1', y='UMAP2', z='UMAP3', color='Library', title="OutputOFplotname")
fig.show()
```

```python
tsne = TSNE(n_components=3, perplexity=30, n_iter=300)
reduced_descriptors_3d = tsne.fit_transform(descriptors_array)
```

```python
df_reduced = pd.DataFrame(reduced_descriptors_3d, columns=['TSNE1', 'TSNE2', 'TSNE3'])
df_reduced['Library'] = df['Library']
```

```python
fig = px.scatter_3d(df_reduced, x='TSNE1', y='TSNE2', z='TSNE3', color='Library', title="OutputOFplotname")
fig.show()
```
